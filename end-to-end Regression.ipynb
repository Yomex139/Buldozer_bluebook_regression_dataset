# modelling for X and Y only

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
Rfr = RandomForestRegressor(n_estimators=40,
                            n_jobs=-1,
                            max_samples=None,
                           random_state=45,
                           min_samples_leaf=1,
                           min_samples_split=14,
                           max_features=0.5)

RFR = RandomForestRegressor(n_jobs=-1,
                            random_state=45,
                            max_samples=None
                             )

grid = {
        'min_samples_leaf': [1,2],
        'min_samples_split': [10,12,14],
        'n_estimators': [20,40,80],
         'max_features': ['auto',0.5,1]}
RandomCV = RandomizedSearchCV(estimator=RFR,
                              param_distributions=grid,
                             cv=5,
                             n_iter=100,
                             n_jobs=-1,
                             random_state=45,
                             verbose=True)
GridCV = GridSearchCV(estimator=RFR,
                      param_grid=grid,
                      cv=5,
                      n_jobs=-1,
                      verbose=True)
```

```python
RFR.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_log_error,r2_score
#Pred_train = 
#Pred_valid = 
model_score = {}
#compute root mean square log error
def RMSLE(y_true,y_pred):
    '''
    return root mean square log error
    '''
    return np.sqrt(mean_squared_log_error(y_true,y_pred))

#evaluation(scoring)
def scoring_evaluation(X_train,X_valid,Y_train,Y_valid,model):
    model.fit(X_train,Y_train)
    score = {'RMSLE_train':RMSLE(Y_train,model.predict(X_train)),
             'RMSLE_valid':RMSLE(Y_valid,model.predict(X_valid)),
             'score_train':model.score(X_train,Y_train),
             'score_valid':model.score(X_valid,Y_valid),
             'r2_score_train':r2_score(Y_train,model.predict(X_train)),
             'r2_score_valid':r2_score(Y_valid,model.predict(X_valid))
            }
    model_score = score
    return model_score


    
```

```python
scoring_evaluation(X_train,X_valid,Y_train,Y_valid,model=
                  Rfr)
```

```python
scoring_evaluation(X_train,X_valid,Y_train,Y_valid,model=
                  RandomCV)
```

```python
%%time
scoring_evaluation(X_train,X_valid,Y_train,Y_valid,model=
                  GridCV)
```
